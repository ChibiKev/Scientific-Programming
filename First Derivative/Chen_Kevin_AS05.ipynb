```python
# Import The Appropriate Libraries
import pandas as pd
import numpy as np

# Data Table Creation
d = {
    'H':[],
    '5 Point Value':[],
    'Exact Value':[],
    'Difference':[]
}
```

```python
x = 1/10 # x = 1/10 radians

for k in range (1, 20, 1): # For multiple values of h
    h = k/20 # h values based on loop
    # five-point midpoint formula using f(x) = sin^2(x)
    fivePoint = (1/(12*h))*((np.sin(x-(2*h))**2)-(8*np.sin(x-h)**2)+(8*np.sin(x+h)**2)-(np.sin(x+(2*h))**2))
    # Actual derivative of the sin^2(x)
    exact = 2*np.sin(x)*np.cos(x)
    d['H'].append(h) # Save
    d['5 Point Value'].append(fivePoint) # Save
    d['Exact Value'].append(exact) # Save
    d['Difference'].append(abs(exact-fivePoint)) # Save
```

```python
# Creates a Table

display(pd.DataFrame(d, columns = ['H','5 Point Value','Exact Value','Difference']))
```

We can see from the table above that as the value of 'H' increases, the value of the '5 Point Value' becomes more distant to the 'Exact Value'. This could also be seen by comparing the 'H' with the 'Difference'. We know this is true because from the lecture "Differentiation" on slide 3, it says that "The simplest way to calculate f'(x0) is to pick a "small" h and calculate"; and from the table, we can see that with the "small" h, it is the simplest way to calculate f'(x0). The reason for this is because it's highly succepible to rounding errors since for "small" h, f(x0+h) can be very close to f(x0).
