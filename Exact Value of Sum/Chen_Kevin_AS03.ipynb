```python
# Kevin Chen
# Collaborated with Simranjit Kaur

# Import The Appropriate Libraries
import math;
import time;
import pandas as pd
import numpy as np

# Data Table Creation
d = {
    'Naive Summation':[],
    'Compensated Summation':[],
    'Exact Value of the Sum':[],
    'Absolute Error (Naive)':[],
    'Relative Error (Naive)':[],
    'Absolute Error (Compensated)':[],
    'Relative Error (Compensated)':[],
    'Elapsed Time (Naive)':[],
    'Elapsed Time (Compensated)':[]
}
```

```python
# Value = sqrt(3), N = 10^5
Value = math.sqrt(3.0)
N = 10**5

# Naive Sum

naive_sum = 0; # Start at 0

start_time = time.time() # Get Start Time
# Calculation Time
for i in range (0,N):
    naive_sum += Value
# Calculation Done
end_time = time.time() # Get End Time

print("Naive Sum: " + str(naive_sum)) # Print Naive Sum
d['Naive Summation'].append(str(naive_sum)) # Store

elapsed_time = end_time - start_time # Get Elapsed Time
print("Elapsed Time (Naive): " + str(elapsed_time) + " seconds") # Print Elapsed Time
d['Elapsed Time (Naive)'].append(str(elapsed_time)) # Store

# Compensated Sum

compensated_sum = 0; # Start at 0
x = 0; # Start at 0

start_time = time.time() # Get Start Time
# Calculation Time
for i in range (0,N):
    temp = compensated_sum
    y = Value + x
    compensated_sum = y + temp
    x = (temp - compensated_sum) + y
    compensated_sum += x
# Calculation Done
end_time = time.time() # Get End Time

print("Compensated Sum: " + str(compensated_sum)) # Print Compensated Sum
d['Compensated Summation'].append(str(compensated_sum)) # Store

elapsed_time = end_time - start_time # Get Elapsed Time
print("Elapsed Time (Compensated): " + str(elapsed_time) + " seconds") # Print Elapsed Time
d['Elapsed Time (Compensated)'].append(str(elapsed_time)); # Store

# Exact Sum
exact_sum = (N) * Value # Calculate
print("Exact Sum: " + str(exact_sum)) # Print Exact Sum
d['Exact Value of the Sum'].append(str(exact_sum)) # Store

# Error for Naive Sum
abs_error_naive = abs(exact_sum - naive_sum) # Calculate
print("Absolute Error (Naive): " + str(abs_error_naive)) # Print Absolute Error
d['Absolute Error (Naive)'].append(str(abs_error_naive)) # Store

rel_error_naive = (abs_error_naive/exact_sum)*100 # Calculate
print("Relative Error (Naive): " + str(rel_error_naive) + " %") # Print Relative Error
d['Relative Error (Naive)'].append(str(rel_error_naive)) # Store

# Error for Compensated Sum
abs_error_compensated = abs(exact_sum - compensated_sum) # Calculate
print("Absolute Error (Compensated): " + str(abs_error_compensated)) # Print Absolute Error
d['Absolute Error (Compensated)'].append(str(abs_error_compensated)) # Store

rel_error_compensated = (abs_error_compensated/exact_sum)*100 # Calculate
print("Relative Error (Compensated): " + str(rel_error_compensated) + " %") # Print Relative Error
d['Relative Error (Compensated)'].append(str(rel_error_compensated)) # Store
```

Theses errors are consistent with the expressions for errors that we discussed in class and that are in the posted classroom presentations. As seen from the table above, as the value of N increases, the absolute error and relative error for the naive way increases. The reason for this increase is because for every additional number that needs to be added to the sum, there would be a small error due to rounding and the error would become noticible when there are more additions. Meanwhile, as the value of N increases, the absolute error and relative error for compensated remains at 0 or low. The reason for this is because for the compensated method, it is more accurate since it reduces rounding errors.
In the speed of the two methods, the naive way seems to be faster than the compensated way. As N increases, we can notice that the compensated way takes longer and longer compared to the naive way. This is mainly because of the multiple computations it has to perform compared to the naive way. Since it's performing more computations, it'll take longer to compute, resulting in a longer time to execute or a slower speed.
