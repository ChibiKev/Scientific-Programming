```python
# Kevin Chen

# Import The Appropriate Libraries
import pandas as pd
import math
from decimal import *

# Data Table Creation
d = {
    'N':[],
    'Value':[]
}
```

```python
x_values = [] # Storage for x

for k in range(17): # k = 0-16
    x = (k*2*math.pi)/16 # Formula for x
    x_values.append(x) # Adding x to x_values

sin_x_values = [] # Storage for sin(x)

for x in range(len(x_values)): # Running through x_value
    if (x%8 == 0): # sin(pi) = 0, python gives you a weird value
        f = 0 # Set f = 0
    else: # Other values are good
        f = math.sin(x_values[x]) # Sin(x)
    sin_x_values.append(f) # Adding f to sin_x_values

# f = sin(0) = 0
# f' = sin'(0) = cos(0) = 1
# f'' = sin''(0) = cos'(0) = -sin(0) = 0
# f''' = sin'''(0) = cos''(0) = -sin'(0) = -cos(0) = -1
# f''''= sin''''(0) = cos'''(0) = -sin''(0) = -cos'(0) = sin(0) = 0 <--- Repeats
sin_derivatives = [0,1,0,-1]

# Tn(x) = summation(k = 0, n, f^(k)(0)*(x)^k /k!)
# Find |Tn(2pik/16)-sin(2pik/16)| < 10^-14, Find Lowest n value
# In x_value = 2pik/16 sin_x_value = sin(2pik/16)

for k in range(len(sin_x_values)): # Running through k
    total = 0 # Currently total = 0
    n = 0 # Start at n = 0
    print("Currently At k =",k)
    while True: # Keep Running Until Conditions are met
        multiply = (Decimal(math.pow(x_values[k],n)))/(Decimal(math.factorial(n))) # Decimal for exact representation
        total += sin_derivatives[n%4] * multiply # Calculating
        print("N =",n,"Calculation:", (abs(total - Decimal(sin_x_values[k])))) 
        if ((abs(total - Decimal(sin_x_values[k]))) < math.pow(10,-14)): # Conditions Met
            d['Value'].append(abs(total - Decimal(sin_x_values[k]))) # Save
            d['N'].append(n) # Save
            break # Next k value
        else:
            n += 1 # Continue if not met
```

```python
# Creates a Table

display(pd.DataFrame(d,columns = ['N','Value'],index=['K = 0','K = 1','K = 2','K = 3','K = 4','K = 5','K = 6','K = 7','K = 8','K = 9','K = 10','K = 11','K = 12','K = 13','K = 14','K = 15','K = 16']))
```

We can see from the table above that as the value of 'k' increases, the value to find the smallest integer 'n' such that |Tn(2*pi*k/16)-sin(2*pi*k/16)| < 10^-14 also increases. The only time it does not increase is when it's k % 8 = 0; this is because the equation makes it so it keeps suming zero and multiplying, which is just zero. This makes sense because from the equations we used, as k increases, the more oscillation there is from the computations, making it take longer to obtain a smaller value.
In addition, the prints above the table shows which 'k' we are working with and the process that 'n' goes through along with the summation of the total. This information is useful in debugging and testing if we are doing the correct commutations; it also shows how we arrived at our answer for the smallest integer 'n'. We can also observe from looking at the prints that the maximum value starts to increase as k increase. The increase of the maximum value would make the oscillation more to obtain a smaller value, making 'n' increase. For example, for k = 1, the maximum value is 0.3826834323650897817792326805, compared to k = 15, the maximum value is 31.30688722654300408058484586.
